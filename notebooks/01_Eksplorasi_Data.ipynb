```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import torch

# Cek GPU dulu, jangan sampe gak kepake
print(f"Status GPU: {'READY ✅' if torch.cuda.is_available() else 'NOT DETECTED ❌'}")

# Load data (kita ambil 100rb baris dulu biar gak lag)
path = "../project_ids/data/raw/train.csv"
print("Sedang memindai dataset...")
df = pd.read_csv(path, nrows=100000)

# Munculin grafik serangan
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='label')
plt.title('Distribusi Serangan (Initial Scan)')
plt.show()

print(f"Total Kolom: {len(df.columns)} fitur terdeteksi.")
```

```python
# 1. Cek Distribusi Label (Serangan vs Normal)
print("Menghitung distribusi serangan...")
label_counts = df['label'].value_counts()
print(label_counts)

# 2. Visualisasi biar 'Manusiawi'
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.barplot(x=label_counts.values, y=label_counts.index, palette='viridis')
plt.title('Distribusi Label pada CIC-IoT2023 (Initial Scan)')
plt.xlabel('Jumlah Data')
plt.ylabel('Jenis Serangan')
plt.show()
```
